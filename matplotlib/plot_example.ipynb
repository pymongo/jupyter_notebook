画动图，动态画图，做到像 turtle chaos 例子那样动态画股票图的方法

> python -m turtledemo.chaos

```python
import numpy as np
import matplotlib.pyplot as plt

i = 3
plt.axis([0, i, 0, i])
xs = [0, 0]
ys = [1, 1]

for index in range(i):
    # rotate
    xs[0], ys[0] = xs[1], ys[1]

    xs[1] = index
    ys[1] = np.random.random() * index

    # 似乎要用 qt/gtk 那样桌面 GUI 才行，网页的 GUI 没法动态画图
    plt.plot(xs, ys)
    # similar to time.sleep(0.8)
    plt.pause(0.8)
```

---
通过 sleep 和 IPython.display 画出会动的图

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn
from IPython.display import display, clear_output
import time
x = np.linspace(0,5)
f, ax = plt.subplots()
ax.set_title("Bessel functions")

for n in range(1,10):
    time.sleep(0.5)
    ax.plot(x, jn(x,n))
    clear_output(wait=True)
    display(f)

# close the figure at the end, so we don't get a duplicate
# of the last plot
plt.close()
```
